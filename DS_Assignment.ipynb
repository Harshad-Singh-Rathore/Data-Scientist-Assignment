Task 2: Merge both the APIs

```python
df_all = pd.merge(
    df_all_predicted,
    df_all_actual,
    on=['time', 'location', 'terrain'],
    how='inner'
)
```

```python
df_all.head()
```

```python
df_all['time'] = pd.to_datetime(df_all['time'])
```

```python
df_all['bias'] = df_all['wind_speed_actual'] - df_all['wind_speed_predicted']
```

```python
df_all['hour'] = df_all['time'].dt.hour
df_all['month'] = df_all['time'].dt.month
```

```python
df_all['wind_dir_sin'] = np.sin(np.radians(df_all['wind_direction_predicted']))
df_all['wind_dir_cos'] = np.cos(np.radians(df_all['wind_direction_predicted']))
```

```python
df_analysis = df_all.copy()
```

```python
df_all = pd.get_dummies(
    df_all,
    columns=['location', 'terrain'],
    drop_first=True
)
```

Task 3: Split the data

```python
train = df_all[df_all['time'] < '2025-11-01'].copy() # Jan-Oct
test  = df_all[df_all['time'] >= '2025-11-01'].copy() # Nov-Dec
```

Task 4: Build a correction model

```python
# Feature Selection
features = [col for col in df_all.columns if col not in [
    'time',
    'wind_speed_actual',
    'bias'
]]
```

```python
X_train = train[features]
y_train = train['bias']

X_test = test[features]
y_test = test['bias']
```

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Linear regression model training
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
bias_pred_lr = lr_model.predict(X_test)
test['corrected_lr'] = test['wind_speed_predicted'] + bias_pred_lr
```

```python
# XG boosting Model Training
xgb_model = XGBRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

xgb_model.fit(X_train, y_train)
```

```python
bias_pred_xgb = xgb_model.predict(X_test)
test['corrected_xgb'] = test['wind_speed_predicted'] + bias_pred_xgb
```

```python
#Evaluation function
def evaluate(true, pred):
    rmse = np.sqrt(mean_squared_error(true, pred))
    mae = mean_absolute_error(true, pred)
    bias = (pred - true).mean()
    return rmse, mae, bias
```

```python
# Raw forecast
rmse_raw, mae_raw, bias_raw = evaluate(
    test['wind_speed_actual'],
    test['wind_speed_predicted']
)

# Linear Regression
rmse_lr, mae_lr, bias_lr = evaluate(
    test['wind_speed_actual'],
    test['corrected_lr']
)

# XGBoost
rmse_xgb, mae_xgb, bias_xgb = evaluate(
    test['wind_speed_actual'],
    test['corrected_xgb']
)
```

```python
#comparison between raw forcast vs linear regressio vs xgboosting
results = pd.DataFrame({
    'Model': ['Raw', 'Linear Regression', 'XGBoost'],
    'RMSE': [rmse_raw, rmse_lr, rmse_xgb],
    'MAE': [mae_raw, mae_lr, mae_xgb],
    'Bias': [bias_raw, bias_lr, bias_xgb]
})
results
```

For Wind Direction

```python
train = df_all[df_all['time'] < '2025-11-01'].copy()
test  = df_all[df_all['time'] >= '2025-11-01'].copy()
```

```python
def signed_angle_diff(forecast, actual):
    return ((actual - forecast + 180) % 360) - 180

df_all["dir_bias"] = signed_angle_diff(
    df_all["wind_direction_predicted"],
    df_all["wind_direction_actual"]
)
```

```python
features = [col for col in df_all.columns if col not in [
    'time',
    'wind_direction_actual',
    'dir_bias'
]]
```

```python
lr_dir = LinearRegression()
lr_dir.fit(train[features], train["dir_bias"])
```

```python
test["dir_corrected_lr"] = (
    test["wind_direction_predicted"] + bias_lr
) % 360
```

```python
xgb_dir = XGBRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

xgb_dir.fit(train[features], train["dir_bias"])
```

```python
bias_xgb = xgb_dir.predict(test[features])
```

```python
test["dir_corrected_xgb"] = (
    test["wind_direction_predicted"] + bias_xgb
) % 360
```

```python
def evaluate_direction(true, pred):
    diff = ((pred - true + 180) % 360) - 180  # signed difference

    rmse = np.sqrt(np.mean(diff**2))
    mae = np.mean(np.abs(diff))
    bias = np.mean(diff)

    return rmse, mae, bias
```

```python
rmse_raw, mae_raw, bias_raw = evaluate_direction(
    test["wind_direction_actual"],
    test["wind_direction_predicted"]
)

# Linear Regression
rmse_lr, mae_lr, bias_lr = evaluate_direction(
    test["wind_direction_actual"],
    test["dir_corrected_lr"]
)

# XGBoost
rmse_xgb, mae_xgb, bias_xgb = evaluate_direction(
    test["wind_direction_actual"],
    test["dir_corrected_xgb"]
)
```

```python
results_dir = pd.DataFrame({
    'Model': ['Raw', 'Linear Regression', 'XGBoost'],
    'RMSE': [rmse_raw, rmse_lr, rmse_xgb],
    'MAE': [mae_raw, mae_lr, mae_xgb],
    'Bias': [bias_raw, bias_lr, bias_xgb]
})
results_dir
```
