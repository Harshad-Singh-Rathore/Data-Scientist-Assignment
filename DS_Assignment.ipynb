Task 2: Merge both the APIs

```python
df_all = pd.merge(
    df_all_predicted,
    df_all_actual,
    on=['time', 'location', 'terrain'],
    how='inner'
)
```

```python
df_all.head()
```

```python
df_all['time'] = pd.to_datetime(df_all['time'])
```

```python
df_all['bias'] = df_all['wind_speed_actual'] - df_all['wind_speed_predicted']
```

```python
df_all['hour'] = df_all['time'].dt.hour
df_all['month'] = df_all['time'].dt.month
```

```python
df_all['wind_dir_sin'] = np.sin(np.radians(df_all['wind_direction_predicted']))
df_all['wind_dir_cos'] = np.cos(np.radians(df_all['wind_direction_predicted']))
```

```python
df_analysis = df_all.copy()
```

```python
df_all = pd.get_dummies(
    df_all,
    columns=['location', 'terrain'],
    drop_first=True
)
```

Task 3: Split the data

```python
train = df_all[df_all['time'] < '2025-11-01']  # Jan-Oct
test  = df_all[df_all['time'] >= '2025-11-01'] # Nov-Dec
```

Task 4: Build a correction model

```python
# Feature Selection
features = [col for col in df_all.columns if col not in [
    'time',
    'wind_speed_actual',
    'bias'
]]
```

```python
X_train = train[features]
y_train = train['bias']

X_test = test[features]
y_test = test['bias']
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from xgboost import XGBRegressor
```

```python
# Model Training
model = XGBRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
bias_pred = model.predict(X_test)

test['corrected_wind'] = (
    test['wind_speed_predicted'] + bias_pred
)
```

Task 5: Evaluation on the test set

```python
#RMSE
rmse_raw = np.sqrt(mean_squared_error(
    test['wind_speed_actual'],
    test['wind_speed_predicted']
))

rmse_corrected = np.sqrt(mean_squared_error(
    test['wind_speed_actual'],
    test['corrected_wind']
))

print(f'RMSE before correction: {rmse_raw}')
print(f'RMSE after correction: {rmse_corrected}')
```

```python
# Mean Absolute Error
mae = mean_absolute_error(
    test['wind_speed_actual'],
    test['corrected_wind']
)
mae
```

```python
# Bias
bias = (test['corrected_wind'] - test['wind_speed_actual']).mean()
bias
```
